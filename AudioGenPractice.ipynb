```python
from google.colab import files
import gradio as gr
import IPython.display as ipd
import numpy as np
```

```python
def process_audio(audio):
    if audio is not None:
        audio_data, sampling_rate = audio
        audio_array = np.frombuffer(audio_data, dtype=np.int16)

        if audio_array.size > 0:
            ipd.display(ipd.Audio(audio_array, rate=sampling_rate))
            ipd.display(ipd.HTML("<p>Audio file processed successfully!</p>"))
        else:
            ipd.display(ipd.HTML("<p>Audio array is empty or invalid.</p>"))
    else:
        ipd.display(ipd.HTML("<p>Unable to process the given audio file.</p>"))
```

```python
audio_input = gr.components.Audio(label="Select an audio file")
output_text = gr.components.Textbox()

interface = gr.Interface(
    fn=process_audio,
    inputs=audio_input,
    outputs=output_text
)
```

```python
interface.launch(share=True, debug=True)
```
